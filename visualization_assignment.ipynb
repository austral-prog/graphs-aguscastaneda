## Grafico 1: Horas de entrenamiento

Este grafico muestra la cantidad de horas de entrenamiento realizadas durante una semana. Elegi un grafico de lineas porque permite visualizar la evolucion de una variable a lo largo del tiempo.

```python
# Ej 1

import numpy as np
import matplotlib.pyplot as plt

dias = ["Lunes", "Martes", "Miercoles", "Jueves", "Viernes", "Sabado", "Domingo"]
horas = [2, 3, 1, 3, 3, 2, 0]

plt.figure(figsize=(8,5))
plt.plot(dias, horas, color='red', marker='o', linestyle='--', label='hs etrenamientos')

plt.title("Horas de entrenamiento durante una semana")
plt.xlabel("Dias de la semana")
plt.ylabel("Horas de entrenamiento")
plt.legend()
plt.grid(True)

plt.savefig("imgs/line_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Grafico 2: Goles por equipo

Este grafico de barras representa la cantidad de goles de distintos equipos. Lo elegi porque es ideal para comparar valores entre diferentes categorias. Las diferencias se observan claramente mediante la altura de cada barra.

```python
# Ej 2

import numpy as np
import matplotlib.pyplot as plt

equipos = ["River", "Boca", "San lorenzo", "Racing", "Independiente"]
goles = [5, 3, 1, 2, 3]

plt.figure(figsize=(8,5))
plt.bar(equipos, goles, color='red')

plt.title("Cantidad de goles de los 5 grandes:")
plt.xlabel("Equipos")
plt.ylabel("Goles")
plt.grid(True)

plt.savefig("imgs/bar.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Grafico 3: Distribución de notas

Este histograma muestra la distribucion de las notas de un conjunto de alumnos. Elegi este tipo de grafico porque permite analizar como se distribuyen los valores dentro de un rango.

```python
# Ej 3

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
notas = np.random.normal(loc=6.5, scale=1.5, size=200)

notas = np.clip(notas, 1, 10)

plt.figure(figsize=(8,5))

plt.hist(notas, bins=10, color='red', edgecolor='black', alpha=0.8)

plt.title("Distribucion de notas de alumnos")
plt.xlabel("Notas")
plt.ylabel("Cantidad de alumnos")

plt.grid(axis='y')

plt.savefig("imgs/histogram.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Grafico 4: Relacion entre estudio y nota

Este grafico de dispersion muestra la relacion entre las horas de estudio y la nota obtenida.

```python
# Ej 4

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
horas_estudio = np.random.uniform(1, 10, 50)
notas = 2 + 0.8 * horas_estudio + np.random.normal(0, 1, 50)

notas = np.clip(notas, 1, 10)

plt.figure(figsize=(8,5))

plt.scatter(horas_estudio, notas, color='black', marker='o', label='Alumnos')


m, b = np.polyfit(horas_estudio, notas, 1)
plt.plot(horas_estudio, m*horas_estudio + b, color = "red", linestyle='--', label='Tendencia')

plt.title("Relacion entre horas de estudio y nota")
plt.xlabel("Horas de estudio")
plt.ylabel("Nota obtenida")
plt.legend()
plt.grid(True)

plt.savefig("imgs/scatter_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Grafico 5: Distribución de lenguajes

Este grafico circular representa la distribucion del conocimiento en distintos lenguajes de programacion.

```python
import matplotlib.pyplot as plt

lenguajes = ["Python", "JavaScript", "C", "ASM", "C#"]
pct = [30, 25, 15, 15, 5]

plt.figure(figsize=(6,6))

plt.pie(pct, labels=lenguajes, startangle=90)

plt.title("Conocimiento de lenguajes")

plt.savefig("imgs/pie_chart.png", dpi=300, bbox_inches='tight')
plt.show()
```
